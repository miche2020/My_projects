```python
# здесь соберем список списков(эл-т - список параметров конкретного автомобиля)
carsURLs = []


def parsing_url(brand):
    cities = ['kazan',  'novosibirsk', 'ekaterinburg']
    for city in cities:
        for i in range(1, 94):     # кол-во страниц для парсинга с авто.ру
            response = requests.get(
                f'https://auto.ru/{city}/cars/{brand}/used/?output_type=list&page={i}')
            print(f'{brand} {city} page={i}')
            with start.get_lock():
                start.value += 1
                print(start.value)
            if response.status_code != 200:
                raise BaseException(
                    "response code" + str(response.status_code))
            soup = BeautifulSoup(response.text, 'html.parser')
            div = soup.select(
                '.ListingItem-module__main .ListingItem-module__thumb a')

            for j in div:
                carsURLs.append(j['href'])
        
    return carsURLs
```

```python
start = Value('I',0)
pool = ThreadPool()
results = pool.map(parsing_url, brand)
```
